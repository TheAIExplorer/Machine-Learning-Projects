To create a model to detect people wearing masks, not wearing them, or wearing masks improperly using the Inception network, you can follow the steps below:

1. Load the dataset and parse the XML files:

```python
import os
import cv2
import xml.etree.ElementTree as ET

dataset_path = "C:/Users/haris/test-installation/Data/Face Mask"
image_folder = os.path.join(dataset_path, "images")
annotation_folder = os.path.join(dataset_path, "annotations")

def parse_annotation(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    filename = root.find("filename").text
    image_path = os.path.join(image_folder, filename)
    image = cv2.imread(image_path)
    
    boxes = []
    labels = []
    
    for obj in root.findall("object"):
        label = obj.find("name").text
        xmin = int(obj.find("bndbox/xmin").text)
        ymin = int(obj.find("bndbox/ymin").text)
        xmax = int(obj.find("bndbox/xmax").text)
        ymax = int(obj.find("bndbox/ymax").text)
        boxes.append((xmin, ymin, xmax, ymax))
        labels.append(label)
    
    return image, boxes, labels

dataset = []
for filename in os.listdir(annotation_folder):
    if filename.endswith(".xml"):
        xml_path = os.path.join(annotation_folder, filename)
        image, boxes, labels = parse_annotation(xml_path)
        dataset.append((image, boxes, labels))
```

2. Preprocess the dataset by resizing the images and encoding the labels:

```python
import numpy as np
from tensorflow.keras.utils import to_categorical

input_size = (224, 224)

def preprocess_dataset(dataset):
    images = []
    labels = []
    
    for image, boxes, labels_list in dataset:
        image = cv2.resize(image, input_size)
        images.append(image)
        
        encoded_label = 0  # Default to "with_mask" if no label is present
        
        if len(labels_list) > 0:
            label = labels_list[0]  # Use the first label as the primary label
            if label == "without_mask":
                encoded_label = 1
            elif label == "mask_worn_incorrectly":
                encoded_label = 2
        
        labels.append(encoded_label)
    
    images = np.array(images)
    labels = np.array(labels)
    labels = to_categorical(labels, num_classes=3)
    
    return images, labels

```

3. Split the dataset into training and testing sets:

```python
from sklearn.model_selection import train_test_split

# Preprocess the dataset
images, labels = preprocess_dataset(dataset)

# Split the dataset into training and testing sets
train_images, test_images, train_labels, test_labels = train_test_split(images, labels, test_size=0.2, random_state=42)
```

4. Build and train the Inception network:

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Dropout

# Data augmentation
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
train_datagen.fit(train_images)

# Fine-tuning the InceptionV3 model
base_model = InceptionV3(weights="imagenet", include_top=False, input_shape=(224, 224, 3))

# Add regularizer
regularizer = l2(0.01)

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation="relu", kernel_regularizer=regularizer)(x)
x = Dropout(0.5)(x)  # Dropout for regularization
predictions = Dense(3, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=predictions)

# Unfreeze some layers for fine-tuning
for layer in model.layers[-30:]:
    layer.trainable = True

# Adjust the learning rate
optimizer = Adam(learning_rate=0.0005)  # Adjust the learning rate accordingly

model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"])

# Add learning rate reduction and early stopping callbacks
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=8, verbose=1)

# Train the model with augmented data
history = model.fit(
    train_datagen.flow(train_images, train_labels, batch_size=32),
    epochs=50,  # Increase the number of epochs for training
    validation_data=(test_images, test_labels),
    callbacks=[reduce_lr, early_stopping]
)
```

5. Evaluate the model on the test set and plot relevant graphs:

```python
import matplotlib.pyplot as plt

# Evaluate the model on the test set
loss, accuracy = model.evaluate(test_images, test_labels)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)

# Save the model
model.save(r'C:\Users\haris\test-installation\Data\Face Mask\Facemask detection.h5')

# Plot accuracy and loss graphs
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
